```python
import os, sys
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## Revalidation

```python
PROJECT_DIR = '../checkpoint/dhs_resnet50_0001'
REVAL_TRAIN_DIR = os.path.join(PROJECT_DIR,'reval_train.csv')
REVAL_DIR = os.path.join(PROJECT_DIR,'reval.csv')
```

```python
xref = np.linspace(-5,5,100)
```

```python
### Training dataset
df_train = pd.read_csv(REVAL_TRAIN_DIR)
y = df_train['y'].tolist()
y0 = df_train['y0'].tolist()
plt.figure()
plt.plot(xref,xref, linewidth=0.5, c='g', alpha=0.3)
plt.scatter(y,y0, s=3, alpha=0.2)
plt.title(f'$r^2$={r2_score(y,y0)}')
```

```python
### Val dataset
df_val = pd.read_csv(REVAL_DIR)
y_val = df_val['y'].tolist()
y0_val = df_val['y0'].tolist()
plt.figure()
plt.plot(xref,xref, linewidth=0.5, c='g', alpha=0.3)
plt.scatter(y_val,y0_val, s=3, alpha=0.2)
plt.title(f'$r^2$={r2_score(y_val,y0_val)}')
```
